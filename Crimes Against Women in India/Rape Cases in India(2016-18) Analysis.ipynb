```python
# State with Highest Number Of Rapes (2016 - 2018)
states = r16["State/UT"].values
c16 = r16["Cases Reported"].values
c17 = r17["Cases Reported"].values
c18 = r18["Cases Reported"].values
total_cases = c16 + c17 + c18

data = [go.Bar(
    x = states,
    y = total_cases,
    marker = dict(color = 'blue',line=dict( width = 2))
)]

layout = go.Layout(
   {
      "title":"State with Highest Number Of Rapes (2016 - 2018)",
   }
)

fig = go.Figure(data=data,layout = layout)
iplot(fig)
```

```python
# TOP 10 States with Highest Number Of Rapes (2016 - 2018) 
states = r16["State/UT"].values
c16 = r16["Cases Reported"].values
c17 = r17["Cases Reported"].values
c18 = r18["Cases Reported"].values
total_cases = c16 + c17 + c18
data = pd.DataFrame({'states': states, 'total_cases': total_cases}, columns=['states', 'total_cases'])
data = [go.Bar(
    x = data.sort_values(by='total_cases',ascending = False)[:10].states.values,
    y = data.sort_values(by='total_cases',ascending = False)[:10].total_cases.values,
    marker = dict(color = 'blue',line=dict(width = 2))
)]

layout = go.Layout(
   {
      "title":"TOP 10 States with Highest Number Of Rapes (2016 - 2018)",
   }
)

fig = go.Figure(data=data,layout = layout)
iplot(fig)
```

```python
# TOP 10 States with Lowest Number Of Rapes (2016 - 2018)
states = r16["State/UT"].values
c16 = r16["Cases Reported"].values
c17 = r17["Cases Reported"].values
c18 = r18["Cases Reported"].values
total_cases = c16 + c17 + c18
data = pd.DataFrame({'states': states, 'total_cases': total_cases}, columns=['states', 'total_cases'])
data = [go.Bar(
    x = data.sort_values(by='total_cases',ascending = True)[:10].states.values,
    y = data.sort_values(by='total_cases',ascending = True)[:10].total_cases.values,
    marker = dict(color = 'blue',line=dict(width=2))
)]

layout = go.Layout(
   {
      "title":"TOP 10 States with Lowest Number Of Rapes (2016 - 2018)",
   }
)

fig = go.Figure(data=data,layout = layout)
iplot(fig)
```

```python
# State with HighestNumber Of Rapes from 2016 to 2018(Year-Wise)
trace1 = go.Bar(x=states, y=c16,name="2016")
trace2 = go.Bar(x=states, y=c17,name="2017")
trace3 = go.Bar(x=states, y=c18,name="2018")

# Fill out  data with our traces
data = [trace1, trace2,trace3]
# Create layout and specify title, legend and so on

layout = go.Layout(title="State with Most Number Of Rapes (2016, 2017, 2018)",
                   xaxis=dict(title="States"),
                   yaxis=dict(title="Number of Cases Reported"),
                   legend=dict(x=1.0, y=0.5)
                   ,barmode="group")


# Create figure with all prepared data for plot
fig = go.Figure(data=data, layout=layout)
# Create a plot in your Python script directory with name "bar-chart.html"
iplot(fig)
```

```python
df = final_df[["State/UT","Cases Reported"]]

df.replace(to_replace='Arunachal Pradesh',value='Arunanchal Pradesh',inplace=True) 

shp_gdf = gpd.read_file('../input/india-gis-data/India States/Indian_states.shp')
merged = shp_gdf.set_index('st_nm').join(df.set_index('State/UT'))

fig, ax = plt.subplots(1, figsize=(12, 12))
ax.axis('off')
ax.set_title('State-wise Reported Rape-Cases (2016-2018)',
             fontdict={'fontsize': '14', 'fontweight' : '4'})
fig = merged.plot(column='Cases Reported', cmap='YlOrBr', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)


```

```python
# Number Of Rape Cases Reported Per Year
c16 = r16["Cases Reported"].values
c17 = r17["Cases Reported"].values
c18 = r18["Cases Reported"].values
total_cases = c16 + c17 + c18
data = [go.Bar(
    x = ["2016","2017","2018"],
    y = [c16.sum(),c17.sum(),c18.sum()],
    marker = dict(color = 'blue',line=dict(width=1.2))
)]

layout = go.Layout(
   {
      "title":"Number Of Rape Cases Reported Per Year",
   }
)

fig = go.Figure(data=data,layout = layout)
iplot(fig)
```

```python
# Age Distribution of Victims
tag = final_df.columns.tolist()[1:-3]
tag.pop(4)
value = [final_df[i].sum() for i in final_df.columns.tolist()[1:-3]]
value.pop(4)
fig = px.pie(values=value,names=tag,
             color_discrete_sequence=px.colors.qualitative.G10)
fig.show()
    
```

```python
# State-Wise Distribution of Age Groups of victims 
for i in tag:
    df = final_df[["State/UT",i]]
    df.replace(to_replace='Arunachal Pradesh',value='Arunanchal Pradesh',inplace=True) 
    shp_gdf = gpd.read_file('../input/india-gis-data/India States/Indian_states.shp')
    merged = shp_gdf.set_index('st_nm').join(df.set_index('State/UT'))
    fig, ax = plt.subplots(1, figsize=(10, 10))
    ax.axis('off')
    ax.set_title('State-wise Rape-Cases Reported (2016-2018) For '+ str(i),
             fontdict={'fontsize': '15', 'fontweight' : '3'})
    fig = merged.plot(column=i, cmap='YlOrBr', linewidth=0.3, ax=ax, edgecolor='0.2',legend=True)
    df.sort_values(by=[i], inplace=True,ascending=False)
    
    fig = px.pie(values=df[i].values,names=df['State/UT'],
             color_discrete_sequence=px.colors.qualitative.G10)
    fig.show()
    
    fig = px.bar(df,x='State/UT',y=i,color_discrete_sequence=['Blue'])
    fig.show()
```
