```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.colors import n_colors
from plotly.subplots import make_subplots
init_notebook_mode(connected=True)
import cufflinks as cf
cf.go_offline()
```

```python
r19 = pd.read_csv('../input/crime-in-india-2019/Rape 2019.csv')
```

```python
r19.head()
```

```python
plt.figure(figsize = (14,12))
sns.color_palette("pastel")
sns.barplot(data= r19, x= 'Rape (Total) ',y='State/UT',palette="deep")
plt.title('Rape Cases Region wise',size = 22)
```

```python
rdiff = r19.groupby(['State/UT'])['Punishment for Rape','Rape (Total) '].sum().reset_index() 
rdiff = rdiff.melt('State/UT', var_name='Difference',  value_name='vals')
plt.figure(figsize = (18,20))
ax = sns.barplot(y = 'State/UT' , x = 'vals',hue = 'Difference' ,data = rdiff ,palette= 'deep')
plt.title("Difference between rape cases and punishment given")
```

```python
rcdy = r19.groupby(['State/UT'])['Rape in Custody - By Police Personnel','Rape in Custody - By Public Servant ','Rape in Custody  - By Member of Armed Forces ','Rape in Custody - By Management/Staff of Jail/Remand Home/Place of Custody ','Rape in Custody  - By Management/ Staff of Hospital'].sum().reset_index() 
rcdy = rcdy.melt('State/UT', var_name='Category',  value_name='vals')
plt.figure(figsize = (18,24))
ax = sns.barplot(y = 'State/UT' , x = 'vals',hue = 'Category' ,data = rcdy ,palette= 'deep')
plt.title("Region wise Custodial Rape",size=20)
```

```python
r_1 = r19.groupby(['State/UT'])['Other Than Custodial Rape- By Relative/Guardian/Teacher/person in position of trust ','Other Than Custodial Rape- On Pregnant Women ','Other Than Custodial Rape - On Women Below 16 years of age ','Other Than Custodial Rape - On Women Incapable of giving Consent ','Other Than Custodial Rape - By Persons in Control or Dominance over women','Other Than Custodial Rape - On Women with Mental or Physical disability'].sum().reset_index() 
r_1 = r_1.melt('State/UT', var_name='Category',  value_name='vals')
plt.figure(figsize = (18,24))
ax = sns.barplot(y = 'State/UT' , x = 'vals',hue = 'Category' ,data = r_1 ,palette= 'deep')
plt.title("Region wise Non-Custodial Rape",size=22)
```

```python
r_2 = r19.groupby(['State/UT'])['Gang Rape','Rape/Gang Rape of Girl Below 12 Yrs (Imprisonment for life till death) ','Rape/Gang Rape of Girl Below 16 Yrs (Imprisonment for life till death) '].sum().reset_index() 
r_2 = r_2.melt('State/UT', var_name='Category',  value_name='vals')
plt.figure(figsize = (18,24))
ax = sns.barplot(y = 'State/UT' , x = 'vals',hue = 'Category' ,data = r_2 ,palette= 'deep')
plt.title("Region wise GangRape",size=22)
```
