```python
# Yearwise Crime rates 
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df1=pd.DataFrame()
for i in crimes:
    df_crimes=df.groupby(['Year'])[i].sum()
    df1[i]=df_crimes
df1
```

```python
# Percentage Contribution of Crimes
a=[]
for i in crimes:
  a.append(df1[i].sum())
a.sort()
plt.figure(figsize=(11,13))
plt.pie(a,labels=crimes,autopct='%1.2f%%',colors=['black', 'gold', 'skyblue', 'coral','pink','cyan','lightgreen'])
plt.tight_layout()
```

```python
df2=df1.T
df2
```

```python
df2["crime"]=df2.index
df2
```

```python
# Crime Rate Comparisions in 2001 and 2014
plt.figure(figsize=(15,6))
df2.groupby("crime")[2001].sum().plot()
df2.groupby("crime")[2014].sum().plot()
plt.legend()
plt.xticks(rotation=90)
```

```python
# State/UT wise Crime Analysis
plt.figure(figsize=(18,8))
plt.title("State/UT wise Crime Rates",size=14)
df.groupby("STATE/UT")["Total"].sum().sort_values(ascending=False).plot.bar()
```

```python
crimes=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls']

df_state=pd.DataFrame()
for i in crimes:
    df_state_crimes=df.groupby(['STATE/UT'])[i].sum()
    df_state[i]=df_state_crimes
df_state["Total"]=df_state.sum(axis=1)
df_state=df_state.sort_values(by="Total",ascending=False)
df_state.reset_index()
df_state
```

### **1. Uttar Pradesh**

```python
df_up=df_state.loc[df_state.index=="uttar pradesh"]
dfupT=df_up.T.drop("Total")
dfupT
```

```python
# Percentage Distribution of Crimes in Uttar Pradesh(2001-2014)
plt.figure(figsize=(13,6)).patch.set_facecolor('lightcyan')
plt.title("Percentage Distribution of Crimes in Uttar Pradesh(2001-2014)",size=14)
plt.pie(dfupT["uttar pradesh"].sort_values(ascending=False),labels=dfupT.index,autopct='%1.2f%%')
plt.tight_layout()
```

```python
# No.of Crimes Cases of each Category in Uttar Pradesh
plt.figure(figsize=(18,12))
plt.title('No.of Crimes Cases of each Category in Uttar Pradesh',size=14)
dfupT.groupby(dfupT.index)["uttar pradesh"].sum().sort_values(ascending=False).plot.bar()
plt.tight_layout()
```

```python
# Year Wise Distribution of Crimes in Uttar Pradesh
crime_up=['Rape','Kidnapping and Abduction','Dowry Deaths',
        'Assault on women with intent to outrage her modesty',
        'Insult to modesty of Women','Cruelty by Husband or his Relatives',
        'Importation of Girls',"Total"]

df_up_year=df.loc[df['STATE/UT']=='uttar pradesh']

df1_up_yearwise=pd.DataFrame()
for i in crime_up:
    df_crimes_up=df_up_year.groupby(['Year'])[i].sum()
    df1_up_yearwise[i]=df_crimes_up
df1_up_yearwise
```

```python
# Year Wise Crime Cases in Uttar Pradesh
plt.figure(figsize=(12,6))
plt.title("Year Wise Crime Cases in Uttar Pradesh",size=14)
sns.barplot(data=df1_up_yearwise,x=df1_up_yearwise.index,y='Total',color="lightblue")
plt.tight_layout()
```

```python
# Timeline Analysis of Different Crime Cases in Uttar Pradesh through Years
plt.figure(figsize=(17,7)).patch.set_facecolor("lightcyan")
plt.title("Time Line of Analyis of Crime Cases in Uttar Pradesh",size=16)
sns.lineplot(data=df1_up_yearwise,palette='gnuplot_r',linestyle="dashdot")
plt.tight_layout()
```
