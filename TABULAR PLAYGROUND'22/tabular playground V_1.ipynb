```python
from tqdm import tqdm
### using columns wise mean
def impute(df, cols):
    for c in tqdm(cols):
        mean = data[cols].mean()
        data[cols] = data[cols].fillna(mean)
        
    return data
        
cols = remove_row_id.columns
mean_data = impute(remove_row_id, cols)
print(mean_data.isnull().any().sum())
    
```

```python
col = smp["row-col"].apply(lambda x:x.split("-"))
```

```python
ans = []
j=0
for row in tqdm(mean_data.iterrows()):
    for i in col[j:]:
        if int(i[0]) == int(row[0]):
            answer = row[1].loc[i[1]]
            ans.append(answer)
        else:
            break
            
        j+=1
            
            

    
```

```python
def row_mean(data):
    mean_row = data.median(axis=1)
    
    return mean_row
        
```

```python
mean = row_mean(remove_row_id)
mean
```

```python
def compute_mean(mean, cols):
    ans = []
    for col in cols:
        ans.append(mean[int(col[0][0])])
        
    return ans

ans = compute_mean(mean, col)
```

```python
submission = pd.DataFrame({
    
    "row-col": smp["row-col"].values,
    "value": ans
    
})
```
