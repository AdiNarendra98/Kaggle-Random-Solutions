```python
df=df.dropna()#Dropping Nan Values
```

```python
X= df.drop('label',axis=1)
```

```python
y= df['label']
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### VOCAB SIZE
voc_size = 5000
```

ONE HOT ENCODING

```python
messages = X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
messages
```

```python
import nltk
import re
from nltk .corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# DATA PREPROCESSING

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
  print(i)
  review = re.sub('[^a-zA-Z]',' ', messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus[1]
```

ONE HOT ENCODING

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
corpus[1]
```

```python
onehot_repr[1]
```

EMBEDDING REPRESENTATION

```python
sent_length=20# Making all sentences of same size =20;can make it different as well.
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[1]
```

MODEL CREATION

```python
embedding_vector_features = 40# Hyperparameter Can be Modified
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))# No of Nuerons Can play
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y) 
```

```python
X_final.shape,y_final.shape
```

```python
# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.3,random_state=42)
```

MODEL TRAINING

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

PERFORMANCE METRICS and ACCURACY

```python
y_pred=model.predict(X_test)
```

```python
y_pred = np.where(y_pred > 0.5,1,0)# Can play with 0.5 this is threshold
# CAN use AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
print("The Accuracy of the Model is:",accuracy_score(y_test,y_pred)*100,"%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

ADDITIONAL METHODS

```python
#USING DROPOUT LAYERS decreases Overfitting
from tensorflow.keras.layers import Dropout
embedding_vector_features = 40# Hyperparameter Can be Modified
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))# No of Nuerons Can play
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
y_pred = np.where(y_pred > 0.6,1,0)
confusion_matrix(y_test,y_pred)
```

```python
print("The Accuracy of the Model with Dropout Layers is:",accuracy_score(y_test,y_pred)*100,"%")
```
