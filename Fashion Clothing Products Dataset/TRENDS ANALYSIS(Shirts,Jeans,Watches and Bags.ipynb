```python
# Using NLP Analysis to find Trends
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
import string

list_stopwords = set(stopwords.words('english'))
```

```python
nlp=df.copy()
```

```python
nlp['ProductName2'] = nlp['ProductName'].str.lower()
nlp['ProductName2'] = nlp['ProductName2'].apply(word_tokenize)
nlp['ProductName2'] = nlp['ProductName2'].apply(lambda x: [word for word in x if word not in list_stopwords])
nlp['ProductName2'] = nlp['ProductName2'].apply(lambda x : [word.translate(str.maketrans('', '', string.punctuation)) for word in x])
nlp['ProductName2'] = nlp['ProductName2'].apply(lambda x : [word for word in x if len(word) > 1])
```

```python
nlp['Description'] = nlp['Description'].str.lower()
nlp['Description'] = nlp['Description'].apply(word_tokenize)
nlp['Description'] = nlp['Description'].apply(lambda x: [word for word in x if word not in list_stopwords])
nlp['Description'] = nlp['Description'].apply(lambda x : [word.translate(str.maketrans('', '', string.punctuation)) for word in x])
nlp['Description'] = nlp['Description'].apply(lambda x : [word for word in x if len(word) > 1])
```
